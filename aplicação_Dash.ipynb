```python
# CÃ³digo definitivo - com reinicializaÃ§Ã£o do kernel
!pip install dash plotly pandas

import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html
from google.colab import files
from google.colab import output
import os

# Mata qualquer processo na porta 8050
os.system("fuser -k 8050/tcp 2>/dev/null")

# Upload do arquivo
print("ğŸ“ FaÃ§a o upload do arquivo:")
uploaded = files.upload()
arquivo = list(uploaded.keys())[0]

# Carregar dados
df = pd.read_csv(arquivo)
print(f"âœ… Arquivo carregado! {len(df)} linhas.")

# Criar app
app = Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1('ğŸ“Š Dashboard E-commerce',
            style={'textAlign': 'center', 'color': '#333'}),

    dcc.Graph(figure=px.histogram(df, x='PreÃ§o', nbins=30,
                                  title='ğŸ’° DistribuiÃ§Ã£o de PreÃ§os')),
    dcc.Graph(figure=px.histogram(df, x='Nota', nbins=20,
                                  title='â­ DistribuiÃ§Ã£o das Notas')),
    dcc.Graph(figure=px.box(df, x='GÃªnero', y='PreÃ§o',
                            title='ğŸ‘• PreÃ§o por GÃªnero')),
    dcc.Graph(figure=px.bar(df['Marca'].value_counts().head(10).reset_index(),
                           x='Marca', y='count',
                           title='ğŸ† Top 10 Marcas'))
])

# Executar com iframe no Colab
print("\n" + "="*50)
print("ğŸš€ Dashboard serÃ¡ aberto abaixo:")
print("="*50)

output.serve_kernel_port_as_iframe(8050, height='1000px')

if __name__ == '__main__':
    app.run(port=8050, debug=False)
```
